```
class Logistic_Regression:
    def __init__(self):
        pass

    def extend(self,X):
        n,k = X.shape
        return np.hstack((X,np.ones((n,1))))

    def init_weights(self, output_size, input_size):
        np.random.seed(42)
        return np.random.normal(0,0.01, (output_size, input_size))

    def sigmoid(self, X):
        return 1./(1+np.exp(-1*self.weights@X.T))

    def get_probability(self,X):
      
        if X.shape[1] != self.weights.shape[1]:
            X = self.extend(X)
        return self.sigmoid(X)

    def get_acc(self, p, Y, threshold=0.65):
        #print(p)
        return np.sum(Y.reshape((Y.shape[0])) == (p>=threshold)[0])/len(Y)

    def fit(self,X,Y,num_epoch=70, lr=0.01):
       X_extended = self.extend(X)
       self.weights = self.init_weights(Y.shape[1], X.shape[1])
       accuracy = []
       losses = []
       for i in range(num_epoch):
          p = self.get_probability(X)
          W_grad = lr*(p-Y.T)@X
          self.weights -= W_grad
          p = np.clip(p, 1e-10, 1 - 1e-10)
          log_loss = -1*np.mean(Y*np.log(p)+(1-Y)*np.log(1-p))
          losses.append(log_loss)
          acc = self.get_acc(p,Y)
          accuracy.append(acc)
       return losses, accuracy
```

```
model = Logistic_Regression()
losses, accuracy = model.fit(X,Y)
accuracy
```
